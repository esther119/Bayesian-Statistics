```python
import pandas as pd 
import numpy as np
import io
  
df = pd.read_csv(io.BytesIO(uploaded['socialmobility.csv']))
print(df)
```

Choose a Dirichlet prior with a uniform distribution over the probability vector parameter of
the multinomial.

```python
import scipy.stats as sts
prior = sts.dirichlet([1 for i in range(len(df))])
prior.alpha
```

Refer to the conjugate prior table on Wikipedia for how to compute posterior hyperparameters of the Dirichlet posterior distribution

```python
df['posterior'] = df['count'] + prior.alpha #calculate conjugated prior 
posterior = sts.dirichlet(df.posterior)
```

What is the posterior 95% probability interval over the probability that a son will become a skilled laborer if his father was an unskilled laborer?

```python
father_unskilled = sts.dirichlet.rvs(df.posterior[df.father == 'unskilled'], size=10000)
father_unskilled
print(f"The 95% CI that a son will become a skilled laborer if his father was an unskilled laborer is {np.percentile(father_unskilled[:, 2], [2.5, 97.5])}") #only the last one is from unskilled to skilled 
```

What is the posterior 95% probability interval over the probability that a father works on a farm if his son works as a professional?

```python
son_professional = sts.dirichlet.rvs(df.posterior[df.son=="professional"], size=10000)
son_professional
print(f"The 95% CI that a father works on a farm if his son works as a professional is {np.percentile(son_professional[:, 0], [2.5, 97.5])}") #only the last one is from unskilled to skilled 
```
